```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

%matplotlib inline
```

```python
#import data
from sklearn.datasets import make_blobs
#--this make blobs of data of chosen number of features and centers.
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                  random_state=101)
```

```python
data[0]
```

```python
data[1]
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1],c=data[1], cmap='plasma')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_  #---predicted labels
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)

ax1.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(data[1],kmeans.labels_))
print(classification_report(data[1],kmeans.labels_))
```
